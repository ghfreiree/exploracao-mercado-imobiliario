# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/millenagena/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio'] # Cria coluna com soma do valor do alguel com valor do condominío
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU'] # Cria coluna com valor anual do apartamento
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com '  + \
                     dados['Quartos'].astype(str) + ' quarto(s) '  + \
                     ' e '  + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'

# A coluna "descrição" é criada com  várias concatenações de strings para criar uma descrição compreensível

dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não") # Cria uma coluna categórica a partir de uma coluna numérica
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False) # Salva os dados completos com as novas colunas em um csv
```
